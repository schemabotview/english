# Cohesion & Discourse

Notebook 03 built strong sentences. But a page of strong sentences can still read like a pile of bricks with no mortar. This notebook is the mortar: how one sentence hands off to the next, and how a run of sentences becomes a single, flowing thought.

Two words to keep apart:

- **Cohesion** — the *surface* links: pronouns, connectors, repeated threads. The visible glue.
- **Coherence** — the *deeper* logic: the ideas actually follow one another. The sense beneath the glue.

You want both, and they're not the same thing — you can glue incoherent sentences together, and you can have a flow so natural it barely needs glue. We'll build the surface tools first, then the logic that has to sit under them. Six moves: the given→new chain, reference, connectors, signposting, and coherence.

## Start from the known, end on the new

The engine of flow is one habit: **begin each sentence with something the reader already knows, and put the new information at the end.** That new piece then becomes the "known" the next sentence starts from — a chain of handoffs.

```text
S1: ...... [ known ] ...... [ NEW ]
                               |
                               v
S2: ............ [ known ] ...... [ NEW ]
                                     |
                                     v
S3: .................. [ known ] ...... [ NEW ]
```

This is the same instinct as end-focus in Notebook 03 (the new, stressed thing goes last) — now stretched across a paragraph. Watch what happens when sentences *don't* hand off:

> **Jolts:** *We added a cache. Recent results are stored by it. Faster loads were what users noticed. A survey confirmed the improvement.* (the start of each sentence lurches to something new)

> **Flows:** *We added a cache. The cache stores recent results, so latency dropped. Users noticed the faster loads — and a survey confirmed the improvement.* (each sentence opens from where the last one ended)

> **The test:** does each sentence *begin* from something the previous one already established? Like stepping stones, place each one where your foot already is — don't make the reader leap.

## Reference & substitution: stop repeating the noun

English dislikes repeating the same noun; it swaps in shorter stand-ins. This is the most common cohesive glue of all.

| device | example |
|---|---|
| pronoun | *the report … **it*** |
| demonstrative + summary noun | *…prices rose. **This increase** worried us.* |
| substitution | *I need a pen — do you have **one**?* / *She passed; he didn't **do so**.* |
| ellipsis (omit the repeat) | *She can sing and he can [ ] too.* |
| the former / the latter, such, the above | *…tea or coffee; **the latter**, please.* |

Repetition → cohesion:

> *The report was late. The report had errors. We sent the report back.*
> *The report was late and full of errors, so we sent it back.*

> **The one trap: dangling reference.** A `this`/`it` with no clear anchor makes the reader hunt. *The server crashed during the demo. **This** was bad* — *this* what? Pin it with a summary noun: ***This crash** was bad.* The rule of thumb: rarely let `this` stand alone — pair it with a noun (`this delay`, `that pattern`, `such cases`).

## Connectors: signs for the logical turns

Connectors (discourse markers) name the **relationship** between two ideas. Each one makes a promise about what's coming — so the wrong one misleads worse than none at all. Grouped by job:

| relationship | markers |
|---|---|
| addition | *moreover, furthermore, in addition, also* |
| contrast | *however, yet, whereas, on the other hand, by contrast* |
| concession | *although, even so, nonetheless, admittedly* |
| cause | *because, since, as* |
| result | *therefore, thus, consequently, hence, so* |
| sequence | *first, then, subsequently, finally* |
| example | *for instance, for example, such as* |
| restatement | *in other words, that is, namely* |

Two traps:

- **Wrong relationship.** *The plan was cheap; **moreover**, it was slow.* — you promised "and another plus," then delivered a minus. You meant *however*.
- **Overload.** A marker on every sentence reads like a checklist. Many sentences connect by content alone (the given→new chain) and need no signpost at all. Use one where the logical turn would otherwise be missed.

Register and punctuation matter too: *however* is formal and takes a comma (*However, …* / *…; however, …*); *but* is neutral. *Thus / hence* are formal; *so* is everyday. Think of connectors as road signs — a *merge*, a *sharp turn*, a *U-turn*. A wrong sign sends the reader off the road.

## Signposting: tell the reader where they are

Zoom out from sentence-joins to the whole text. **Signposting** tells the reader the shape of what's coming and where they currently stand, so they never wonder *why am I being told this?*

- **Forecast a structure:** *There are three reasons. First… Second… Finally…*
- **Hand off between sections:** *Having covered the costs, let's turn to the risks.* / *So far we've seen X; now consider Y.*
- **Flag the key point:** *The crucial trade-off is this:* … / *What matters here is …*
- **Topic sentences:** open each paragraph with the one idea it will deliver; the rest of the paragraph pays it off. A reader skimming only your first sentences should still get the argument.

> *Buried:* a paragraph that lists details and reveals its point only at the end.
> *Signposted:* the same paragraph led by *The launch was a disaster* — now every detail lands as evidence for a claim the reader already holds.

Signposting is generosity: you do the navigation so the reader spends their attention on the content, not on figuring out the map.

## Coherence: glue can't fix broken logic

Everything above was **cohesion** — surface links. **Coherence** is the deeper test: do the ideas genuinely follow one another? You can have one without the other:

> *cohesion without coherence* — *Sales fell. **Therefore**, the office repainted the lobby. **However**, Tuesday is bin day.* Perfectly glued; says nothing.

So coherence has its own requirements, beneath the connectors:

- **One paragraph, one controlling idea.** Everything in it serves that idea; whatever doesn't belongs in another paragraph or nowhere.
- **A consistent thread.** The topics (often the grammatical subjects) stay related sentence to sentence. Wild subject-hopping breaks the sense even when the connectors are all "correct."
- **Real progression.** *The shuffle test:* if you can reorder the sentences with no loss, the paragraph isn't an argument — it's a list. Genuine flow has a direction you can't scramble.

The lesson: when a paragraph feels off, don't reach for *however* to patch it. The order of *ideas* is usually what's wrong. **Fix the thinking first, then the linking.**

## Try it: join these up

Each item is choppy, repetitive, or mis-linked. Repair it with the named tool. One is done for you.

> **Given (remove repetition):** *The bug was rare. The bug was hard to reproduce. The bug took us a week.*
> **Joined:** *The bug was rare and hard to reproduce, so it took us a week.*
> *(merged with reference "it" and the connector "so")*

Now you:

1. Fix the handoff (given→new) — *Sales rose 20%. A new ad campaign was the cause. Younger buyers were reached by the campaign.*
2. Choose the right connector — *The plan was cheap. ____ it was slow.*
3. Pin the dangling reference — *The server crashed during the demo. This was embarrassing.*
4. Add a topic sentence — *We were late. The slides wouldn't load. Then the mic died.*
5. Spot the break — does this cohere? *The team is small. Therefore, our coffee is excellent.*

Compare with one possible version:

| # | one possible repair |
|---|---|
| 1 | Sales rose 20%, driven by a new ad campaign that reached younger buyers. |
| 2 | The plan was cheap, **but** it was slow. (contrast — not *moreover*) |
| 3 | …crashed during the demo. **This crash** was embarrassing. |
| 4 | **The presentation fell apart.** We were late, the slides wouldn't load, then the mic died. |
| 5 | No — *therefore* promises a cause that isn't there; team size doesn't explain the coffee. Fix the logic, not the link. |

Yours needn't match — if the reader's eye now glides from one sentence to the next, the mortar is holding.
