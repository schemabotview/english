# Precision & Word Choice

At B level you already know enough words to say almost anything. The C-level jump is not *more* words — it is reaching for the *exact* one. Precision is what makes writing feel sharp and speech feel sure of itself; vagueness is what leaves correct English feeling flat.

This notebook works on the **form** layer from Notebook 01 — zoomed all the way in to the single word. Five levers, from the feeling a word smuggles in down to deleting the words that earn nothing:

- **Connotation** — the attitude a word carries.
- **Collocation** — the partners a word expects.
- **Strong verbs** — freeing the verb trapped inside a noun.
- **Cutting filler** — removing words that add nothing.
- **Register** — matching the word to the room.

## Connotation: words carry an attitude

Every word has a **denotation** — its dictionary meaning — and a **connotation** — the feeling and judgement riding along with it. Synonyms share the denotation and split on the connotation, so choosing between them is really choosing a *stance*.

All of these mean "thin":

| word | what it leaks |
|---|---|
| `slim`, `slender` | attractive, deliberate |
| `lean` | fit, athletic |
| `underweight` | clinical, neutral |
| `skinny` | slightly negative, too thin |
| `gaunt` | unwell, hollow |

Same body; five different opinions about it. The reader hears the opinion whether you intended it or not — so the skill is making sure the word leaks the attitude you *actually* hold.

More pairs that trip people up — each shares a meaning but flips the judgement:

| approving | disapproving |
|---|---|
| confident | arrogant |
| assertive | pushy |
| thrifty | stingy |
| curious | nosy |
| determined | stubborn |
| childlike | childish |

> **Everyday image:** two shirts cover you equally well, but one says *job interview* and the other says *beach*. A word can fit the meaning and still wear the wrong message.

## Collocation: words travel in packs

Some words simply expect certain partners. The pairing isn't a grammar rule — it's just what sounds *native*. Get the word right but the partner wrong and a fluent listener flinches, even though nothing is technically "incorrect."

| native | off (but "correct") |
|---|---|
| `make` a mistake | ~~do~~ a mistake |
| `do` your homework | ~~make~~ your homework |
| `heavy` rain | ~~strong~~ rain |
| `strong` coffee | ~~powerful~~ coffee |
| `commit` a crime | ~~make~~ a crime |
| `pay` attention | ~~give~~ attention |
| `take` a risk | ~~do~~ a risk |

This is the **single biggest tell** of an advanced-but-non-native speaker: the grammar is flawless and the word-partnerships are a half-step off.

Intensifiers collocate too — you can't swap them freely:

| set phrase | not |
|---|---|
| `highly` likely | ~~very~~ likely (weaker) |
| `bitterly` disappointed | ~~strongly~~ disappointed |
| `deeply` concerned | ~~very~~ concerned (weaker) |
| `fully` aware | ~~totally~~ aware (informal) |
| `utterly` ridiculous | ~~very~~ ridiculous (weaker) |

How to build this: **collect chunks, not single words.** Store `make a decision` as one unit, the way you once stored a single word — that's how natives carry it.

## Rescue the verb trapped in the noun

A quiet drag on advanced writing: take a perfectly good verb, freeze it into a noun, then prop it up with a weak helper — *make, do, give, conduct, provide, reach*. The result is longer, colder, and slower.

| frozen (noun-heavy) | thawed (strong verb) |
|---|---|
| make a decision | decide |
| conduct an investigation | investigate |
| give consideration to | consider |
| reach a conclusion | conclude |
| provide assistance | help |
| carry out an analysis | analyse |
| is indicative of | indicates |
| put forward a proposal | propose |

> *We will conduct a review of the data and reach a decision.* — 11 words
> *We will review the data and decide.* — 7 words, same meaning, more drive

The verb is where a sentence's energy lives; burying it in a noun freezes it. This isn't a ban — sometimes the action genuinely *is* the topic (*the decision was unpopular*). Overuse is the problem, not the form. The habit to build: whenever you write *make / do / give / conduct + noun*, check whether the noun is a verb in hiding.

## Cut the words that earn nothing

Precision is as much subtraction as selection. **The test: delete the word; if the meaning survives, it was filler.**

**Intensifier inflation.** `very` + a weak word is weaker than one exact word:

| inflated | exact |
|---|---|
| very angry | furious |
| very tired | exhausted |
| very big | enormous |
| very small | tiny |
| very important | crucial |
| very surprised | astonished |

(And `very unique` or `very perfect` are nonsense — those words are already absolute.)

**Filler nouns** — `thing`, `stuff`, `aspect`, `factor`, `area`. A specific noun is almost always hiding behind them: *the financial side* beats *the money aspect*.

**Dead phrases** — long wind-ups that say nothing:

| bloated | tight |
|---|---|
| in order to | to |
| due to the fact that | because |
| at this point in time | now |
| in the event that | if |
| has the ability to | can |
| it is important to note that | (just say it) |

None of these make you sound advanced — they make the reader wait. (Deliberate, careful *hedging* is a genuine skill — that's Notebook 04. This section is about the **accidental** padding, not the chosen kind.)

## Match the word to the room

English carries two vocabularies stacked on top of each other: a short, blunt, warm **Anglo-Saxon** core, and a longer, cooler, more formal **Latin / French** layer that arrived later. Most everyday ideas have one of each:

| plain (Anglo-Saxon) | formal (Latinate) |
|---|---|
| buy | purchase |
| start | commence |
| end | terminate |
| help | assist |
| ask | enquire |
| need | require |
| get | obtain |
| enough | sufficient |

Here's the trap — and it's the most common B→C mistake: **reaching for the fancy word to sound advanced.** It backfires. Over-Latinate writing sounds pompous, distant, even evasive — *We will commence the procedure to obtain the required authorisation* — while plain words carry warmth and trust: *We'll start getting the approval we need.*

Higher register isn't better; it's just *higher*. Mastery is to **default to the plain word and reach for the formal one only when the situation truly calls for it** — a legal notice, a careful complaint, a formal report. Knowing both, and choosing on purpose, is the whole game. (Notebook 06 takes this from single words up to the tone of a whole message.)
