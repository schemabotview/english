# Idioms, Phrasal Verbs & Figurative Language

Everything so far has made your English *correct, precise, well-shaped, and well-pitched*. This notebook adds the last thing: making it sound **native** rather than assembled. The mark of figurative language is that its meaning **isn't the sum of its words** — *bite the bullet* has nothing to do with bullets — and English is saturated with it. Strip it out and your English is flawless yet somehow foreign; the textbook never quite leaves.

A fair warning up front: **this is the layer where overreach backfires hardest.** A misused idiom or a mangled metaphor jars far more than the plain sentence it replaced — it announces "non-native" louder than a small grammar slip. So the goal here is *natural*, not *more* — which is why two of the five parts are about restraint.

Five parts: phrasal verbs, idioms, metaphor & simile, sounding un-translated, and knowing when to hold back.

## Phrasal verbs: the everyday workhorse

A phrasal verb is a verb plus a particle (*give up*, *put off*, *run into*), and it's the engine of natural spoken English — natives reach for it far more than its single formal cousin. The particle changes everything:

| phrase | meaning |
|---|---|
| give **up** | quit |
| give **in** | surrender |
| give **out** | distribute / stop working |
| give **away** | donate / reveal |
| give **off** | emit |

One verb, five meanings — set entirely by the particle.

**Register (Notebook 06):** the phrasal verb is the neutral/spoken option; its single Latinate verb is the formal twin (Notebook 02's two vocabularies) — *put off / postpone*, *find out / discover*, *look into / investigate*, *go up / increase*. Speech leans phrasal; formal writing leans Latinate.

> *Stiff:* *I will investigate the issue and we can discuss it.*
> *Natural (spoken):* *I'll look into it and we can talk it over.*

**One grammar trap — the pronoun rule.** With separable phrasal verbs, a pronoun must sit *inside*: *turn **it** off* ✓, ~~turn off it~~ ✗. (A full noun can go either side: *turn off the light* / *turn the light off*.)

> **Strategy:** don't memorise lists. Learn a handful of high-frequency ones deeply, in real sentences. Depth beats breadth — *get*, *put*, *take*, *come*, and *go* alone spawn dozens you'll actually use.

## Idioms: use only what you've truly heard

An idiom is a fixed expression whose meaning isn't its literal words — *bite the bullet* = do the unpleasant-but-necessary thing. Stick to **high-frequency, living** ones (skip museum pieces like *raining cats and dogs*):

| idiom | meaning |
|---|---|
| play it by ear | decide as you go |
| on the same page | in agreement |
| touch base | check in briefly |
| a grey area | unclear, undefined |
| back to the drawing board | start over |
| a blessing in disguise | a hidden upside |
| cut corners | do it cheaply/poorly |
| get to the point | stop stalling |

Three rules govern them:

- **They're frozen.** You can't swap words: it's *spill the beans*, not ~~spill the peas~~; *the ball's in your court*, not ~~your field~~.
- **They carry register** — mostly informal and spoken; rare in formal writing (Notebook 06).
- **They're culture-bound** — many come from sport or sailing and may baffle international listeners.

> **The big trap: half-remembering.** A mangled idiom is *worse* than plain words — it flags non-native loudly. *A blessing in disguise* becomes *~~a blessing in the sky~~*; *for all intents and purposes* becomes *~~for all intensive purposes~~*. The rule is strict: **use only idioms you've actually heard, kept intact.** Unsure of a single word? Don't use it. An idiom is an inside joke a whole language shares — natural when you land it exactly, awkward the moment you fumble the punchline.

## Metaphor & simile: making the abstract concrete

Figurative comparison turns something abstract into something you can picture.

- **Simile** — an *explicit* comparison with *like* or *as*: *it landed like a punch*, *busy as a bee*.
- **Metaphor** — an *implicit* one, saying X simply *is* Y: *time is money*, *she's a rock*, *a flood of emails*.

You already swim in **dead (conventional) metaphors** without noticing — and that's good, it's just normal English: *grasp* an idea, a deadline *looming*, *drowning* in work, the *heart* of the matter, a *key* point. Native English is built from these.

**Fresh** metaphors add vividness and voice (Notebook 08 turns them into a tool of persuasion), but they carry more risk. Two dangers:

- **Mixed metaphors** — two clashing images in one breath: *~~we'll burn that bridge when we come to it~~*, *~~it's not rocket surgery~~*. Open one picture and finish it in the *same* picture — one image per sentence.
- **Clichés** — metaphors worn smooth: *think outside the box*, *low-hanging fruit*, *move the needle*. So overused they now signal empty corporate noise. Fresh or plain both beat tired.

## Sounding un-translated

Why does correct English still sometimes sound foreign? Usually because it was *translated* word-by-word from a first language, or *assembled* logically, rather than reached for as a ready-made native chunk.

- **Idioms don't port.** A literal translation of your first-language expression produces something English speakers simply don't say. The image that's vivid at home lands as a riddle here.
- **Natives store chunks, not words** (Notebook 02's collocation, again): *make a decision*, *take a break*, *by and large*, *as far as I'm concerned*. They retrieve whole phrases, not individual bricks.
- **The test:** *does a native actually say this, in this exact combination?* If you built it by logic but never heard it, be suspicious.

> **The path is input, not rules.** There's no formula for what sounds natural — only exposure plus attention. Read and listen to real English, notice the chunks that keep recurring, store them whole, and reuse them intact. **Recognition comes before production:** you should have *heard* an idiom many times before you ever deploy it. This layer is grown, not memorised.

## When *not* to reach for it

Because overreach costs most here, restraint is its own skill. Hold back when:

- **The register is formal** (Notebook 06). Idioms and slangy phrasal verbs are wrong in a legal notice, a formal report, or an academic paper — those want the plain or Latinate twin. *We'll look into it* in chat; *we will investigate the matter* in the report.
- **The audience is international.** With non-native listeners, idioms *reduce* clarity. Plain English reaches more people; dropping the idiom is the considerate, not the lazy, choice.
- **You're not 100% sure.** If the exact form or meaning is shaky, use the plain version. A clean plain sentence always beats a broken idiom.
- **You're already using one.** One apt idiom lands; a paragraph stuffed with them reads like a phrasebook showing off. Idiom is *seasoning, not the meal*.

> **The principle:** idiom should feel *inevitable*, not *inserted*. If you're reaching for one mainly to sound advanced — the same trap Notebooks 02 and 06 flagged — that's the moment to choose the plain word instead.

## Try it: make these natural — and know when not to

A mix of tasks. One is done for you.

> **Given (make it natural — phrasal):** *I will postpone the meeting and investigate the matter.*
> **Natural:** *I'll put the meeting off and look into it.*
> *(swapped Latinate verbs for high-frequency phrasal verbs — fits spoken register)*

Now you:

1. Make it sound natural (phrasal) — *Could you distribute these handouts and then return the originals?*
2. Fix the mangled idiom — *For all intensive purposes, it was a blessing in the sky.*
3. Fix the mixed metaphor — *We'll cross that bridge when we burn it.*
4. Choose an idiom that fits — *We don't have a firm plan; let's just decide as we go.*
5. Spot the register problem — *The contract shall be terminated if the client cuts corners.* (formal legal sentence)

Compare with one possible version:

| # | one possible answer |
|---|---|
| 1 | Could you hand these out and then give the originals back? |
| 2 | For all intents and purposes, it was a blessing in disguise. |
| 3 | We'll cross that bridge when we come to it. (one image, kept whole) |
| 4 | Let's play it by ear. |
| 5 | Idiom clash with legal register — replace *cuts corners* with *fails to meet the agreed standard*. Keep idioms out of formal contracts. |

Yours needn't match — if it sounds like something a native would actually say, *and* fits the room, you've got it.
